```python
import sys
import pandas as pd
from states import states_names, state_codes,get_state_code
```

```python
file_path_2016 = 'correctional_populations/cpus16at01.csv'
file_path_2015 = 'correctional_populations/cpus15at01.csv'
file_path_2014 = 'correctional_populations/cpus14at01.csv'
file_path_2013 = 'correctional_populations/cpus13at01.csv'
```

```python
def read_in_csv(filename, year):
    names = ['remove','region','total_correctional_pop','csrp_100k_18','csrp_100k_all','remove1','parole',
             'comsrp_100k_18','comsrp_100k_all','remove3','local_jail_prison','irp_100k_18','irp_100k_all']
    df = pd.read_csv(filename, skiprows=12,skipfooter=10,sep=",", names=names)
    df.drop(df.columns[5], axis=1, inplace=True)
    df.drop(df.columns[8], axis=1, inplace=True)
    df.drop(df.columns[0], axis=1, inplace=True)
    df["year"] = [year for num in range(0,len(df))]
    df['status_code'] = df.apply(get_state_code, axis=1)
    return df

data2016 = read_in_csv(file_path_2016, 2016)
data2016.head()
```

```python
data2015 = read_in_csv(file_path_2015, 2015)
data2015.head()
```

```python
data2014 = read_in_csv(file_path_2014, 2014)
data2014.head()
```

```python
data2013 = read_in_csv(file_path_2013, 2013)
data2013.head()
```

```python
union = pd.concat([data2013, data2014])
union = pd.concat([union, data2015])
union = pd.concat([union, data2016])
union = union[union['total_correctional_pop'].notna()]
union = union[union['region'].notna()]

def to_int(f):
    try:
        return int(str(f).replace(',', ''))
    except:
        return 0
    
def to_all_to_int(df):
    list_names = ['total_correctional_pop','csrp_100k_18','csrp_100k_all','parole','comsrp_100k_18','comsrp_100k_all',
                  'local_jail_prison','irp_100k_18','irp_100k_all']
    for name in list_names:
        df[name] = df[name].map(to_int)
    return df

union = to_all_to_int(union)
union.head()
```

```python
union.to_csv("2013-2016_incarceration_usa.csv")
```
